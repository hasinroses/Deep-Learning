60000 is the number of samples and 28, 28 is the dimension of the individual images

```python
import IPython
url = 'https://raw.githubusercontent.com/hasinroses/DL_Assignment/main/networkimage1.png'
IPython.display.Image(url, width = 550)
```

Convert 28x28 into Single dimentional array which is called flattening

```python
X_train.shape
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train_flattened=X_train.reshape(len(X_train),28*28)
X_test_flattened=X_test.reshape(len(X_test),28*28)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened.shape
```

```python
X_train_flattened[0]
```

```python
model=keras.Sequential([      #sequential means stack of layer of neural network
  keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')   #   Dense মানে হ'ল একটি Layer er Neuron পরের Layer এর সাথে Connect থাকবে'
                                              #এখানে 10 হ'ল Output উপরের ছবিটার মত'
]) 
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train_flattened,y_train,epochs=5) # number of iteration which our neural Network is going to training
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
plt.matshow(X_test[1])
```

```python
y_predict=model.predict(X_test_flattened)
```

```python
y_predict[1] #Shows 10 because we set Output=10
```

```python
np.argmax(y_predict[1])
```

```python
y_predict_labels=[np.argmax(i) for i in y_predict]
y_predict_labels[:5]
```

```python
y_test[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predict_labels)
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

The above model was 2 layer. Now I want to add One hidden layer

```python
model=keras.Sequential([     
  keras.layers.Dense(100,input_shape=(784,),activation='relu'), # ETa Hidden Layer, Ekhane 100 diyesi amar icche moto. 
  keras.layers.Dense(10,activation='sigmoid') # Eta Output Layer tai input hobe na
]) 
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train_flattened,y_train,epochs=5) # number of iteration which our neural Network is going to training
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_predict_2layer=model.predict(X_test_flattened)
y_predict_2layer
```

```python
y_predict_2layer_labels=[np.argmax(i) for i in y_predict_2layer]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predict_2layer_labels)
cm
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predictions')
plt.ylabel('Truth')
```

# In the above example  X_train and  X_test transform to Flatten. It can automatically make Flatten by using keras. See the below example

```python
model=keras.Sequential([     
    keras.layers.Flatten(input_shape=(28,28)),  
    keras.layers.Dense(100,activation='relu'), # Jehetu agei Input Shape bola hoyeche tai ekhane lagbe na
  keras.layers.Dense(10,activation='sigmoid') 
]) 
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train,y_train,epochs=5) 
```
